#### <span style="color:red"> Plain vanilla option pricing
<span style="color:red"> Black-Scholes theory assumes that the time evolution of the asset price is stochastic and exhibits geometric Brownian motion (GBM) with constant risk-free interest rate $r$ and volatility $\sigma$.

<span style="color:red">Black-Scholes analytic solution for the price at time t<T: \
<span style="color:red">- European Call-option price: $$ C\left[ S(t),t \right] = S(t) N(d_1) -K e^{-r(T-t)} N(d_2) $$
<span style="color:red">- European Put-option price: $$ P\left[ S(t),t \right] = S(t) \left[ N(d_1)-1\right] -K e^{-r(T-t)} \left[ N(d_2)-1\right]$$

<span style="color:red"> where 
$$d_1 = \frac{1}{\sigma\sqrt{T-t}}\left[ \ln \left( \frac{S(t)}{K} \right) + \left( r+\frac{\sigma^2}{2} (T-t) \right)\right]$$

<span style="color:red"> where $d_2 = d_1 - \sigma\sqrt{T-t}$, and where
$$ N(x) = \int_{-\infty}^{x} dz \, e^{-\frac{z^2}{2}} = \frac{1}{2} \left[ 1+ {\rm erf} \left( \frac{x}{\sqrt{2}}\right) \right]$$
is the cumulative of Gaussian $\sim \mathcal{N}(\mu=0,\sigma^2=1)$,
and where
$${\rm erf}(x) = \frac{2}{\sqrt{\pi}}\int_{0}^{x} dz \, e^{-z^2}$$

<span style="color:red"> By using the following parameters: 
- <span style="color:red">asset price at $t=0$: $S(0)=100$
- <span style="color:red">delivery time: $T=1$
- <span style="color:red">strike price: $K=100$
- <span style="color:red">risk-free interest rate: $r=0.1$
- <span style="color:red">volatility: $\sigma=0.25$

<span style="color:red">compute at time $t=0$ **via Monte Carlo** the European call-option price, $C[S(0),0]$, and put-option price, $P[S(0),0]$:

1. <span style="color:red"> By sampling **directly** the final asset price $S(T)$ for a $GBM(r,\sigma^2)$
2. <span style="color:red"> By sampling the discretized $GBM(r,\sigma^2)$ path of the asset price
dividing $[0,T]$ in $10^2$ time intervals: $S(0)$, $S(t_1)$, $S(t_2)$, ... up to $S(t_{100}=T=1)$

<p style="border:2px; border-style:solid; border-color:#F5F5F5; padding: 1em; background-color:#F5F5F5">
    Remember that given an asset price $S\sim GBM(\mu,\sigma^2)$, with initial value $S(0)$, then
    $$S(t) = S(0) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)t+\sigma W(t)\right]$$
    where $W(t)\sim\mathcal{N}(0,t)$.
    Moreover, since the increments of $W$ are independent and normally distributed, the simple recursive procedure for sampling discretized values of $S\sim GBM(\mu,\sigma^2)$ at $t_1 < t_2 < … < t_n$ is:
$$S(t_{i+1}) = S(t_i) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)(t_{i+1}-t_i)
    +\sigma Z_{i+1}\sqrt{t_{i+1}-t_i}\right]$$
    with $Z_i \sim \mathcal{N}(0,1)$
</p>

### Processo di Wiener
Il moto Browniano standard o processo di Wiener $W(t),t\geq 0$ è un processo stocastico che soddisfa le seguenti proprietà:
1. $W(0)=0$
2. Gli incrementi $W(t)-W(s)$ sono stazionari e indipendenti
3. Per $t>s$, gli incrementi $W(t)-W(s)$ seguono una distribuzione normale $\mathcal{N}(\mu;\sigma^2)=\mathcal{N}(0;t-s)$

Le proprietà elencate sopra implicano che $W(t)$ è distribuito secondo $\mathcal{N}(0,t)$. Il processo ha media nulla e varianza che cresce linearmente con il tempo $\Rightarrow$ è un processo di diffusione. Per simulare una traiettoria si suddivide un intervallo di tempo in sottointervalli di ampiezza $\Delta t$. Ad ogni passo si genera un incremento $W(t_{i+1})=W(t_i)+Z_{i+1}\sqrt{t_{i+1}-t_i}$ con $Z_{i+1}$ variabili gaussiane standard indipendenti. Un processo di Wiener è autosimilare, che significa che ogni porzione finita di $W(t)$, se propriamente riscalata, è statisticamente indistinguibile dall'intero percorso.

Il calcolo stocastico permette di trattare traiettorie continue ma non differenziabili (come quelle di $W(t)$
), definendo un'integrazione rispetto a un processo aleatorio. Si può dimostrare che la forma del differenziale totale di un funzionale del processo di Wiener è data da:

$$
df[W(t),t]=\left[\frac{\partial f}{\partial t}+\frac{1}{2}\frac{\partial^2 f}{\partial W^2}\right ]dt+\frac{\partial f}{\partial W}dW(t)
$$
che ha la forma di un'equazione differenziale stocastica.

### Equazioni differenziali stocastiche
Un processo stocastico $x(t)$ soddisfa un'**equazione socastica in forma di Itô** (SDE) $dx(t)=a[x(t),t]dt+b[x(t),t]dW(t)$ quando, $\forall t_0,t$ si ha $x(t)=x(t_0)+\int_{t_0}^tdt'a[x(t'),t']+\int_{t_0}^tdW(t')b[x(t'),t']$, i.e. si usa l'integrale stocastico per definire la SDE. $a[x(t),t]$ è la componente deterministica, $b[x(t),t]$ è il coefficiente di diffusione che determina l’intensità della parte casuale, mentre $dW(t)$ è l’incremento infinitesimo di un processo di Wiener.

Si suppone che $x(t)$ soddisfi la SDE precedente e che $f$ sia un funzionale di $x$ e $t$. Facciamo un'espansione di Taylor al secondo ordine per calcolare il differenziale totale. Si ottiene la **formula di Itô**:
$$
df[x(t),t]=\left\{\frac{\partial f[x(t),t]}{\partial t}+a[x(t),t]\frac{\partial f[x(t),t]}{\partial x}+\frac{1}{2}b^2[x(t),t]\frac{\partial^2 f[x(t),t]}{\partial x^2}\right\}dt+b[x(t),t]\frac{\partial f[x(t),t]}{\partial x}dW(t)
$$


### Generazione di processi stocastici
Un processo $X(t)$ è un **moto Browniano** con drift $\mu$ è coefficiente di diffusione $\sigma^2$ ($X\sim BM(\mu,\sigma^2)$, $\sigma>0$ ) se $BM(0,1)\sim \frac{X(t)-\mu t}{\sigma}$ (i.e. è un moto Browniano standard). Segue che $X(t)=\mu t+\sigma W(t)$ e $X(t)\sim \mathcal{N}(\mu t,\sigma^2 t)$ e $X(t)$ risolve la SDE $dX(t)=\mu dt +\sigma dW(t)$. 

Dato che il moto Browniano ha incrementi indipendenti e distribuiti normalmente, è molto semplice da simulare il processo: si considerano $Z_1,\dots,Z_n$ variabili casuali indipendenti distribuite normalmente. Per $BM(\mu,\sigma^2)$, dato $X(0)=X_0$, i valori successivi possono essere generati tramite 
$$X(t_{i+1})=X(t_i)+\mu(t_{i+1}-t_i)+\sigma Z_{i+1}\sqrt{t_{i+1}-t_i}$$
Si tratta di un metodo stocastico nel senso che la distribuzione congiunta dei valori simulati coincide con la distribuzione congiunta del corrispondente moto Browniano.

### Modelli di mercati finanziari
Un mercato è un "luogo" dove venditori e acquirenti si trovano per scambiare dei "prodotti". Un prodotto in un certo istante di tempo $t$ ha un costo $S(t)$ detto spot price, che varia a seconda di domanda/richiesta.

Si definisce un **mercato efficiente** se gli operatori ottengono rapidamente tutte le informazioni rilevanti, se è liquido (i.e. si può comprare e vendere facilmente) e ha bassa frizione (i.e. bassi costi di transazione). Se il mercato è efficiente i prezzi attuali contengono tutta l’informazione passata. Si considera che, se si potesse prevedere un aumento futuro del prezzo, tutti gli investitori acquisterebbero subito, ma un mercato efficiente risponderebbe immediatamente all'aumentata richiesta con un aumento del prezzo, annullando così l'opportunità di guadagno. Da ciò segue che le correlazioni nei prezzi durano un tempo molto breve e il processo di fluttuazione dei prezzi risulta casuale $\Rightarrow$ è sensato fare un'**ipotesi di Markov**, secondo cui l'evoluzione futura del mercato dipende solo dallo stato attuale e non da tutta la storia passata.

Per studiare l'evoluzione di $S(t)$, si assume che al tempo $t=0$ depositiamo una somma $S(0)$ in una banca, che garantisce un tasso di interessi senza rischio $r$ per il deposito.
$$
dS(t)=\mu S(t)dt+\sigma S(t) dW(t)
$$
Dove il **drift** $\mu$ misura il tasso di crescita medio dell'asset price $S(t)$ e la **volatilità** $\sigma$ misura l'intensità delle fluttuazioni statistiche di prezzo.

La SDE definisce una variante del moto Browniano, detto **moto Browniano geometrico**, in cui $a[.]$ e $b[.]$ sono proporzionali alla variabile casuale $S(t)$. In questo caso la variabile importante non è tanto $dS=S(t+dt)-S(t)$, ma $dS/S(t)$. Usando la formula di Itô si trova:
$$
d\ln{S(t)}=(\mu-0.5\sigma^2)dt+\sigma dW(t)
$$
La soluzione di questa SDE è:
$$
S(t)=S(0)\exp\left[(\mu-\frac{1}{2}\sigma^2)t+\sigma W(t)\right]
$$
In alternativa si può simulare una traiettoria con GBM usando
$$
S(t_{i+1})=S(t_i)\exp \left[ (\mu-\frac{1}{2}\sigma^2)
(t_{i+1}-t_i)+\sigma Z_{i+1}\sqrt{t_{i+1}-t_i}\right]
$$


L'**opzione europea** è contratto tra due parti in cui colui che vende l'azione, il **writer**, detiene un obbligo, e chi acquista l’opzione, l'**holder**, detiene un diritto. L’opzione attribuisce all’holder il diritto, ma non l’obbligo, di comprare (call option) oppure vendere (put option) l’attività sottostante, a un prezzo fissato $K$, detto **strike price**, in una data futura $T$ predeterminata. Se $S(t)>K$ all'holder conviene comprare l'azione, altrimenti è conveniente venderla. Poiché il writer si assume un rischio potenziale, al momento della stipula richiede un premio d'opzione, cioè un pagamento iniziale che rappresenta il costo per ottenere il diritto descritto nel contratto.

Il principale problema diventa capire quanto dovrebbe costare un opzione. Black e Scholes rispondono a questa domanda per:
- mercato massimamente efficiente (infinitamente liquido e senza frizione)
- L'evoluzione dell'asset price segue un moto Browniano geometrico.

Nell'esercizio si simula l'andamento del valore dell'asset price $S(t)$ usando:
- **simulazione diretta:** si campiona direttamente il prezzo finale usando
$$
S(T)=S(0)\exp \left[ (\mu-\frac{1}{2}\sigma^2)T+\sigma Z\sqrt{T}\right]
$$

- **simulazione discreta:** si campiona l’asset price con un percorso discretizzato del GBM, suddividendo l’intervallo $[0,T]$ in un numero finito di passi, e aggiornando il prezzo a ogni tempo tramite $$
S(t+\Delta t)=S(0)\exp \left[ (\mu-\frac{1}{2}\sigma^2)\Delta t+\sigma Z\sqrt{\Delta t}\right]
$$

Il profitto alla scadenza è dato da $\text{max}[0,S(t)-K]$. Per ottenere il valore attuale di questo profitto al tempo $t$ bisogna scontarlo di un fattore $e^{-rt}$ dovuto agli interessi. Il calcolo dell'option price corrisponde al calcolo del valore di aspettazione
$$
E[e^{-rt}\text{max}[0,S(t)-K]]=\langle e^{-rt}\text{max}[0,S(t)-K]\rangle
$$
che è una stima di quanto l'holder deve pagare al tempo $t=0$ sulla base del profitto atteso al tempo $t=T$.


Sono state effettuate $M=10000$ simulazioni Monte Carlo, suddivise in $N=100$ blocchi da $L=100$ simulazioni ciascuno. L’analisi degli errori statistici è stata condotta tramite la tecnica del data blocking, calcolando medie progressive e relative incertezze. I risultati ottenuti con i due metodi sono compatibili entro gli errori statistici e convergono al valore teorico fornito dalla formula di Black–Scholes, confermando la correttezza dell’implementazione e la validità dell’approccio Monte Carlo.


```python
import numpy as np
import matplotlib.pyplot as plt

# ================================
#    CARICAMENTO dati CALL
# ================================
dataC = np.loadtxt("outputC.txt")
nC, meanC, errC = dataC[:,0], dataC[:,1], dataC[:,2]

BS_call = 14.975790778311286

# ================================
#    CARICAMENTO dati PUT
# ================================
dataP = np.loadtxt("outputP.txt")
nP, meanP, errP = dataP[:,0], dataP[:,1], dataP[:,2]

BS_put = 5.4595325819072364

# ================================
#       FIGURA AFFIANCATA
# ================================
fig, ax = plt.subplots(1,2, figsize=(13,5))  # NOTA: niente sharey=True

# ---- CALL ----
ax[0].errorbar(nC, meanC, yerr=errC, fmt='.-', color='slateblue', alpha=0.7)
ax[0].axhline(BS_call, color='red', linestyle='-', label="Analitico B&S")
ax[0].set_title("European CALL - Sampling diretto")
ax[0].set_xlabel("blocco")
ax[0].set_ylabel("Prezzo")
ax[0].grid(True)
ax[0].legend()

# ---- PUT ----
ax[1].errorbar(nP, meanP, yerr=errP, fmt='.-', color='slateblue', alpha=0.7)
ax[1].axhline(BS_put, color='red', linestyle='-', label="Analitico B&S")
ax[1].set_title("European PUT - Sampling diretto")
ax[1].set_xlabel("blocco")
ax[1].set_ylabel("Prezzo")
ax[1].grid(True)
ax[1].legend()

plt.suptitle("European Option Pricing – Convergenza Monte Carlo", fontsize=13)
plt.tight_layout()
plt.show()
```
