<span style="color:red">In Statistical Mechanics the one-dimensional (1D) Ising Model is exactly resoluble. The Hamiltonian for a system of N spins with nearest neighbors interaction is given by (in the following I will assume units such that $\mu_B=1$ and $k_B=1$):
$$
H = 
-J\sum_{i=1}^N s_i s_{i+1}
-\frac{h}{2}\sum_{i=1}^N (s_i + s_{i+1}) 
$$

<span style="color:red">where $\{ s_i \}_{i=1,N}$ are spin variables taking values in $\{-1;1\}$, $J>0$ (ferromagnetic model),and with the assumption $s_{N+1} = s_{1}$.

<span style="color:red">The partition function for $N$ spins, $Z$, can be exactly computed and it turns out to be:
$$
Z=\sum_{\{ s_i \}} \exp \left( -\beta H\right) = \lambda_1^N + \lambda_2^N
$$

<span style="color:red">where
$\lambda_{1,2} = 
e^{\beta J} 
\cosh (\beta h) \pm \left[ e^{2\beta J} \cosh^2 (\beta h) - 2\sinh (2\beta J) \right]^{1/2}$, 
and where $\beta=\frac{1}{k_b T} = \frac{1}{T}$.
Once the partition function has been computed, the Helmholtz free energy is readly obtained as $A = -k_B T \ln Z$. All the thermodynamic quantities of interest can be deduced from $A$ or, equivalently, from $Z$:

- <span style="color:red">the **Internal Energy** per spin (with $h=0$):
$$
\frac{U(N,T)}{N} = \frac{\langle H \rangle}{N}
= \frac{1}{N} \sum_{\{ s_i \}} H \frac{\exp \left( -\beta H\right)}{Z}
= -\frac{1}{N}\frac{\partial \ln Z}{\partial \beta}
\underset{h=0}= -J \tanh (\beta J) \Biggl( \frac{1 + \tanh^{N-2} (\beta J)}{1+\tanh^N (\beta J)} \Biggl)
$$
- <span style="color:red">the **Heat Capacity** (with $h=0$):
$$
C(N,T) = \frac{\partial U(N,T)}{\partial T}
= \frac{\partial \beta}{\partial T}\frac{\partial U}{\partial \beta}
= -k_B\beta^2 \frac{\partial}{\partial \beta} \left[ \sum_{\{ s_i \}} H \frac{\exp \left( -\beta H\right)}{Z} \right] = \\
= -k_B\beta^2 \left[ -\sum_{\{ s_i \}} H^2 \frac{\exp \left( -\beta H\right)}{Z} + \sum_{\{ s_i \}} H \exp \left( -\beta H\right)\frac{\partial Z^{-1}}{\partial \beta} \right] = \\
= k_B\beta^2 \left\{ \sum_{\{ s_i \}} H^2 \frac{\exp \left( -\beta H\right)}{Z} + \left[ \sum_{\{ s_i \}} H \frac{\exp \left( -\beta H\right)}{Z}\right]\frac{1}{Z}\frac{\partial Z}{\partial \beta} \right\} =\\
= k_B\beta^2 \left\{ \sum_{\{ s_i \}} H^2 \frac{\exp \left( -\beta H\right)}{Z} - \left[ \sum_{\{ s_i \}} H \frac{\exp \left( -\beta H\right)}{Z}\right]^2 \right\} =
k_B\beta^2 \left(\langle H^2 \rangle -\langle H \rangle^2 \right)
$$
<span style="color:red"> $$
\frac{C(N,T)}{N} = \frac{1}{N}\frac{\partial U(N,T)}{\partial T} =\\
\underset{h=0}= 
k_B (\beta J)^2 \left\{ 1+\tanh^N (\beta J)+(N-1)\tanh^2 (\beta J)+(N-1)\frac{\tanh^{N-2} (\beta J)}{1+\tanh^N (\beta J)}
- N \left[ \frac{\tanh (\beta J) + \tanh^{N-1} (\beta J)}{1+\tanh^N (\beta J)} \right]^2 \right\}
$$
- <span style="color:red">The **magnetization** (with $h \not= 0$):
Note that with $h=0$ the magnetization $M(N,T,h=0)=\langle \sum_{i=1,N}s_i \rangle =0$ because in 1D we cannot have a (ferromagnetic) phase transition in absence of an external magnetic field. ( $h$ è il campo esterno )
$$
M(N,T,h) = -\frac{\partial A}{\partial h} = k_B T \frac{\partial \ln Z}{\partial h} =
\frac{1}{\beta} \frac{1}{Z}\frac{\partial}{\partial h}\sum_{\{ s_i \}} \exp \left( -\beta H\right) = \sum_{\{ s_i \}} \left[ \sum_{i=1,N}s_i\right] \frac{\exp \left( -\beta H\right)}{Z} = \left\langle \sum_{i=1,N}s_i \right\rangle
$$
The exact formula from $M(N,T,h) = k_B T \dfrac{\partial \ln Z}{\partial h}$ is:
$$
\frac{M(N,T,h)}{N}=\frac{e^{\beta J}\; \sinh(\beta h)}{Z} \; \Bigl[\lambda_1^{N-1}\cdot \Bigl(1+\frac{e^{\beta J}\; \cosh(\beta h)}{\sqrt{e^{2\beta J} \cosh^2 (\beta h) - 2\sinh (2\beta J)}}\Bigr) + \lambda_2^{N-1}\cdot \Bigl(1-\frac{e^{\beta J}\; \cosh(\beta h)}{\sqrt{e^{2\beta J} \cosh^2 (\beta h) - 2\sinh (2\beta J)}}\Bigr)\Bigr]
$$

- <span style="color:red"> The **magnetic susceptibility** (with $h=0$):
$$
\chi(N,T) = \frac{\partial M}{\partial h} = 
\frac{\partial}{\partial h}\sum_{\{ s_i \}} \left[ \sum_{i=1,N}s_i\right] \frac{\exp \left( -\beta H\right)}{Z} =\\
= \beta\sum_{\{ s_i \}} \left[ \sum_{i=1,N}s_i\right]^2 \frac{\exp \left( -\beta H\right)}{Z}
-\sum_{\{ s_i \}} \left[ \sum_{i=1,N}s_i\right] \frac{\exp \left( -\beta H\right)}{Z}
\frac{1}{Z}\frac{\partial Z}{\partial h} = 
\beta \left[ \left\langle \left( \sum_{i=1,N}s_i \right)^2 \right\rangle - \left\langle \sum_{i=1,N}s_i \right\rangle^2 \right] = \\
\underset{h=0}= \beta \left[ \left\langle \left( \sum_{i=1,N}s_i \right)^2 \right\rangle \right] = \beta e^{2\beta J} \Biggl( \frac{1-\tanh^N (\beta J)}{1+\tanh^N (\beta J)} \Biggl)
$$

### Modello di Ising
In questa esercitazione consideriamo il <span style="color:red">modello di Ising</span>, ovvero un modello semplificato per descrivere il magnetismo nella materia. L'idea alla base è quella di rappresentare magnete come un reticolo, di geometria qualsiasi, in d dimendioni (d=1, 2, 3). In ogni sito del reticolo si trova un dipolo magnetico o spin. 

Il momento angolare totale $\vec{j}$ è legato al momento di dipolo magnetico $\vec{\mu}$ nel seguente modo: 

$$
\vec{\mu}=g\frac{e}{2mc}\vec{j}, \hspace{1cm} \mu^2=g^2\mu_B^2j(j+1), \hspace{1cm} \mu_z=g\mu_B m \hspace{1cm} m=-j,-j+1,\dots,j-1, j
$$

Dove
- $\mu_B=\frac{eh}{2mc}$ è il magnetone di Bohr
- $g$ è il fattore di Landè

Osservazioni teoriche e sperimentali hanno mostrato che, per molti <span style="color:red">materiali ferromagnetici</span>, il momento angolare totale è compatibile un un valore di $\vec{j}=1/2$. In questo caso $g=2$ e vale la semplificazione:
$$
 \mu^2=g^2\mu_B^2s(s+1) \hspace{1cm} \mu_z=\pm \mu_B 
$$

Dove $s$ è il numero quantico associato allo spin dell'elettrone $\vec{s} = 1/2$, $s=\pm 1/2$.

Noi ci concentriamo sul caso monodimensionale, <span style="color:red">la catena di spin</span>, in cui abbiamo $N$ spin disposti su una circonferenza, equispaziati. Si assume che l'interazione avvenga soltanto tra primi vicini, con la stessa intensità per ogni coppia scelta. Sotto queste ipotesi, l'Hamiltoniana del sistema è data da (in unità t.c. $\mu_B=1$ and $k_B=1$):

$$
H = 
-J\sum_{i=1}^N s_i s_{i+1}
-\frac{h}{2}\sum_{i=1}^N (s_i + s_{i+1}) 
$$

Dove
- $J$ è una costante che descrive la tendenza degli elettroni ad allinearsi paralleli o antiparalleli
- $\{ s_i \}_{i=1,N}$ sono le variabili di spin che possono assumere valore -1 o +1 (in questa notazione $s_{N+1} = s_{1}$)
- $h$ è il campo magnetico esterno applicato.
  
Per d=1,2 esiste una soluzione esatta del modello di Ising, quindi saremo in grado di confrontare la simulazione con la soluzione analitica.

### Metodo Montecarlo applicato al modello di Ising

Si osservi che per una catena di $N$ spin come quella descritta il sistema può assumere $2^N$ configurazioni. Vorremmo simulare un sistema di Ising di taglia finita usando metodi Monte Carlo, in modo tale da poter fare una stima dei valori a ogni data temperatura $k_BT=1/\beta$. Il peso statistico di Boltzmann per uno stato generico $\mu$ è data da 

$$
p(\mu)=\frac{e^{-E_\mu/k_BT}}{\sum_{\nu=1}^{2^N}e^{-E_\nu/k_BT}} \hspace{1cm} \Rightarrow \hspace{1cm} \langle G \rangle = \sum_\mu g(\mu)p(\mu)
$$

Si parte da una configurazione iniziale $\mu_0$, i.e. si scelgono una temperatura e la direzione dello spin per ciascuna particella (up o down). Il punto di partenza non modifica il risultato finale in modo significativo, ma alcune scelte sono migliori di altre. Noi partiamo da una configurazione a temperatura infinita: in questa situazione l'energia dovuta all'interazione tra primi vicini è trascurabile rispetto all'energia termica delle particelle, che può essere usata per modificare la direzione dello spin. Di conseguenza gli spin saranno orientati up o down in modo casuale. Un'altra configurazione iniziale possibile è quella a $T=0$, in cui il sistema si trova nel ground state: con tutti gli spin allineati o up o down (in caso di campo magnetico esterno nullo le due configurazioni sono equivalenti. In presenza di campo esterno gli spin si allineano parallelamente al campo magnetico per materiali ferromagnetici e antiparallelamente per materiali antiferromagnetici).

Il secondo step consiste nel proporre una mossa: si genera uno stato $\nu$ con una probabilità di transizione di prova $T(\mu|\nu)=T(\nu|\mu)$ e si accetta la mossa con probabilità:

$$
A(\nu|\mu)=\min\left[1,\frac{T(\mu|\nu)p(\nu)}{T(\nu|\mu)p(\mu)}\right]=\min\left[1,\frac{p(\nu)}{p(\mu)}\right]=\min\left[1,e^{-\beta(E_\nu-E_\mu)}\right]
$$

Per semplificare il codice si calcola $\Delta E=2Js_k^\mu\sum_{i n.n. to k}s_i^\mu$:
- se $\Delta E>0$ genero un numero casuale $0\leq n_{rand}<1$, se $n_{rand}<A(\mu|\nu)$ accetto, altrimenti rifiuto la mossa.
- se $\Delta E<0$ accetto la mossa.

Si osserva che, con questo metodo, la funzione di partizione si semplifica nell'accettanza, così è possibile campionare la distribuzione di probabilità e calcolare i valori medi senza conoscerla.

In parole semplici, si sceglie in modo casuale uno spin alla volta nella catena e si prova a ruotare lo spin. 
Uno step di Metropolis finisce dopo si è provato a flippare tutti gli $N$ spin nella catena. È importante non seguire un ordine nella scelta dello spin da flippare per evitare correlazioni.





### Metodo di Gibbs applicato al modello di Ising

Il metodo di Gibbs è un metodo di campionamento che può essere visto come un caso particolare dell'algoritmo di Metropolis, in cui ogni mossa proposta viene accettata. 
Nel modello di Ising si sceglie una configurazione iniziale e si visita ogni sito o in ordine o in modo casuale e si sceglie un nuovo valore di spin, indipendente da quello attuale, dalla distribuzione condizionata definita dagli altri n.n. spin.

Nel modello di Ising, ogni spin $s_i=\pm 1$ ha una probabilità che dipende dai vicini:

$$
p(s_i|vicini)=\frac{1}{1+e^{-2\beta J \sum_j s_j}}
$$
	​


Il campionamento di Gibbs funziona così:

1. Si fanno scorrere tutti gli spin del reticolo.
2. Per ogni spin $s_i$, si calcola la probabilità condizionata che lo spin valga +1, dato lo stato attuale dei vicini.
3. Si estrae $s_i$ da una Bernoulli con probabilità calcolata.
4. Si ripete lo stesso procedimento per tutti i siti, questo costituisce un passo di Gibbs.
5. Dopo molti passi, la configurazione di spin segue la distribuzione di Boltzmann: $p(\{s_i\})\propto e^{-\beta E(\{s_i\})}$



### Equilibrazione
Come prima cosa in realtà si esegue un processo di equilibrazione del sistema: lo scopo della simulazione è quello di calcolare i valori medi delle quantità di interesse. Per farlo dobbiamo conoscere la distribuzione di probabilità, che campioniamo grazie all'algoritmo Metropolis o Gibbs. Il problema è che durante i primi passi della simulazione l'algoritmo non campiona bene la distribuzione di probabilità, ma il campionamente diventa buono asintoticamente. L'equilibrazione serve per fare una stima di quanto tempo (dove per tempo si intende numero di step) serve per raggiungere questo limite. Una volta stimato il numero di step necessari $N_{eq}$, si calcolano i valori medi
partendo dallo step $N_{eq}+1$, in modo tale da rimuovere l'errore sistematico dovuto al funzionamento dell'algoritmo.

$$
\langle G \rangle \cong \frac{1}{N-N_{eq}}\sum_{N_{eq+1}}^N G(\mu_i)
$$

Per stimare quanti step bisogna fare prima di poter calcolare le medie si eseguono $10^4$ passi di Metropolis o Gibbs suddivisi in altrettanti blocchi di singolo passo
