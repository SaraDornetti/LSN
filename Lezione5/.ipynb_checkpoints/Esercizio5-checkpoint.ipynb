<span style="color:red">In quantum physics a **wave function**, $\Psi$, is a mathematical description of the state of a quantum system. The wave function is a complex-valued probability amplitude, and the probabilities for the possible results of measurements made on the system can be derived from it.

<span style="color:red">For now on, let's consider the simple case of a non-relativistic single particle, without spin, in three spatial dimensions. The state of such a particle is completely described by its wave function, $\Psi(\vec{r},t)$, where $\vec{r}$ is position and $t$ is time.

<span style="color:red">For one spinless particle, if the wave function is interpreted as a probability amplitude, the square modulus of the wave function, $|\Psi(\vec{r},t)|^2$, is interpreted as the probability density that the particle is at $\vec{r}$ at time $t$. Once we have a probability density, we can use Monte Carlo ...

#### <span style="color:red">Hydrogen atom
<span style="color:red">The wave functions of the eigenstates of an electron in a Hydrogen atom (this is the only atom for which the Schroedinger equation has been solved exactly) are expressed in terms of spherical harmonics and generalized Laguerre polynomials. It is convenient to use spherical coordinates, and the wave function can be separated into functions of each coordinate:
$$
\Psi_{n,l,m}(r,\theta,\phi)=
\sqrt{\left(\frac{2}{na_0}\right)^3 \frac{(n-l-1)!}{2n[(n+l)!]}}
e^{-r/na_0}\left(\frac{2r}{na_0}\right)^l L_{n-l-1}^{2l+1}\left(\frac{2r}{na_0}\right)
Y_l^m(\theta,\phi)
$$
where $a_0=4\pi\epsilon_0\hbar^2/m_e e^2=0.0529$ nm is the Bohr radius, $L_{n-l-1}^{2l+1}$ are the generalized Laguerre polynomials of degree $n-l-1$, $n=1,2,...$ is the principal quantum number, $l=0,1, ..., n-1$ the azimuthal quantum number, $m=-l, -l+1, ..., l-1, l$ the magnetic quantum number.

<span style="color:red">For example, the ground state wave function is:
$$
\Psi_{1,0,0}(r,\theta,\phi)=
\frac{a_0^{-3/2}}{\sqrt{\pi}} e^{-r/a_0} 
$$
whereas one of the three $2p$ excited state is:
$$
\Psi_{2,1,0}(r,\theta,\phi)=
\frac{a_0^{-5/2}}{8}\sqrt{\frac{2}{\pi}} r e^{-r/2a_0} \cos(\theta)
$$
<span style="color:red">Expectation values for the radius turns out to be exactly:
$$
\left\langle r \right\rangle_{\Psi_{1,0,0}} = \frac{3}{2}a_0
\quad
\left\langle r \right\rangle_{\Psi_{2,1,0}} = 5a_0
$$
</span>

### Algoritmo di Metropolis $M(RT)^2$
L’algoritmo sfrutta una catena di Markov per campionare una distribuzione di probabilità target. Si tratta di un algoritmo basato su una tecnica di rigetto, semplice ed efficiente, che permette di campionare qualunque distribuzione di probabilità, anche in uno spazio a un numero arbitrario di dimensioni, senza la necessità di conoscere la costante di normalizzazione della distribuzione. Il principale svantaggio di questa tecnica è che il campionamento è corretto solo asintoticamente. Inoltre, le variabili successive generate dall’algoritmo possono risultare fortemente correlate tra loro.

L’algoritmo nasce da un’analogia con il comportamento dei sistemi in meccanica statistica, nei quali l’evoluzione del sistema è governata da una matrice di transizione stocastica $K(x|y)$ (generalmente assunta come nota), che rappresenta la probabilità che il sistema passi dallo stato $y$ allo stato $x$. Lasciando evolvere il sistema per un tempo sufficientemente lungo, esso raggiunge l’equilibrio, seguendo una certa distribuzione di probabilità $p(x)$, che però è ignota e che si desidera determinare. L’algoritmo di Metropolis procede in modo opposto: si sceglie a priori la distribuzione $p(x)$ che si vuole campionare e si costruisce una probabilità di transizione $K(x|y)$ tale che $p(x)$ sia una distribuzione invariante rispetto a essa. La dinamica del sistema è quindi governata da $K(x|y)$, e il meccanismo di accettazione delle nuove configurazioni è progettato in modo da garantire che la distribuzione $p(x)$ rimanga invariata sotto l’azione di tale operatore di transizione.

Dato un vettore $x$, $M(RT)^2$ costruisce un RW i cui passi sono progettati in modo t.c., se ripetuti molte volte, la distribuzione asintotica di $x$ sia proprio la distribuzione di probabilità che vogliamo campionare. Supponiamo che $x_1,x_2,\dots$ siano step in un RW, e a ogni $x_i$ è associata una probabilità $d_i(x_i)$, noi vorremmo $$\lim_{n\to\infty}d_n(x)=p(x)$$

Per costruire un RW di questo tipo si parte da un processo di Markov stazionario, in cui la probabilità condizionata (matrice stocastica di trasferimento) non dipende dal tempo $p_{1|1}(x_2,t_2|x_1,t_1)=K(x_2|x_1)$. Il processo è caratterizzato da $p(x)=\int K(x|y) p(y)dy$. Questa equazione integrale mostra che un processo di Markov stazionario ha almeno una distribuzione di probabilità che è invariante rispetto alla probabilità di transizione. Considerando una probabilità di transizione diversa $d_1(x)$, non necessariamente invariante, si costruisce un processo di Markov non stazionario caratterizzato da una sequenza di distribuzioni marginali $d_{n+1}(x)=\int K(x|y) d_n(y)dy$, dove la speranza è che $d_n$ si trasformi velocemente in $p$, per farlo bisogna trovare una $K(x|y)$ adeguata.

Il problema di trovare $K(x|y)$ è abbastanza complicato, perchè bisogna trattare l'equazione integrale $p(x)=\int K(x|y) p(y)dy$. Il problema si risolve usando un trucco: il principio del bilancio dettagliato $K(x|y)p(y)=K(y|x)p(x)$ implica l'equazione integrale che vorremmo risolvere, per cui conviene lavorare con questa equazione. Si riscrive  $K(x|y)= T(x|y)\times A(x|y)$ dove $T(x|y)$ è una probabilità di transizione test, e $A(x|y)$ la probabilità di accettazione. Scegliendo 
$$A(x|y)=\text{min}[1,q(x,y)] \qquad q(x,y)=\frac{T(y|x)p(x)}{T(x|y)p(y)}\geq0$$
il principio del bilancio dettagliato è verificato e segue che $p(x)$ è invariante sotto $K(x|y)$ così definita e che $d_{n+1}(x)=\int K(x|y)d_n(y)dy \to p(x)$ per qualsiasi $d_1(x)$.

Per comodità assumiamo che $T(x|y)=T(y|x)$, per cui $A(x|y)=\text{min}[1,p(x)/p(y)]$.

Nel pratico si sceglie un punto di partenza e si propone una mossa usando una distribuzione di probabilità (non importa quale), che viene accettata o meno usando il meccanismo illustrato sopra.
- Se si propongono mosse troppo grandi si rischia di rifiutare troppe mosse e di avere efficienza molto bassa.
- Se si propongono mosse troppo piccole si accettano quasi tutte le mosse e si rischia di avere una correlazione troppo alta.

La regola empirica è di mantenere l'accettanza intorno al $50\%$ per evitare questi due casi limite.

### Algoritmo di Metropolis applicato all'atomo di idrogeno
L’algoritmo implementa il campionamento Metropolis per stimare il valore medio del raggio $r = \sqrt{x^2 + y^2 + z^2}$ di un elettrone nell’atomo di idrogeno, considerando le densità di probabilità associate agli orbitali $1s$ e $2p$. In entrambi i casi viene campionata direttamente la distribuzione $|\psi(x)|^2$, senza la necessità di conoscere la costante di normalizzazione. L’evoluzione del sistema avviene tramite una catena di Markov nello spazio tridimensionale. A ogni passo Monte Carlo viene proposta una nuova configurazione $x' = x + \Delta x$, dove lo spostamento $\Delta x$ è estratto secondo una distribuzione uniforme nell’intervallo $[-\text{passo}, +\text{passo}]$ per ciascuna coordinata oppure secondo una distribuzione gaussiana centrata in zero con deviazione standard pari a $\text{passo}$. La configurazione proposta viene accettata con probabilità $\alpha = \min\left(1,\frac{p(x')}{p(x)}\right)$,
dove $p(x)$ è la densità di probabilità dell’orbitale considerato.

I parametri utilizzati sono:

Per l’orbitale $1s$ la posizione iniziale è $x_0 = (10,-8,4)$.
- per distribuzione uniforme $\text{passo} = 1.22 \rightarrow \text{acc} = 0.50$
- per distribuzione gaussiana $\text{passo} = 0.69 \rightarrow \text{acc} = 0.53$
  
Per l’orbitale $2p$ la posizione iniziale è $x_0 = (-12,23,-5)$.
- per distribuzione uniforme $\text{passo} = 2.8\rightarrow \text{acc} = 0.52$
- per distribuzione gaussiana $\text{passo} = 1.75\rightarrow \text{acc} = 0.52$

Per ciascuna simulazione viene eseguita una fase di termalizzazione di $10^4$ passi Monte Carlo, seguita da una fase di misura composta da
$M = 10^6$ passi totali, suddivisi in $N = 100$ blocchi. Le medie e le incertezze statistiche sul valore di $\langle r \rangle$ sono calcolate mediante il metodo del data blocking. Dai risultati si osserva che per l'equilibrazione sono sufficienti pochi passi prima che il valore del raggio si stabilizzi intorno a un valore medio. Lo stesso risultato si osserva dai plot 3D in cui si vede che la fase di "esplorazione" prima di iniziare a campionare correttamente la distribuzione di probabilità desiderata dura pochi passi.

```python
import numpy as np
import matplotlib.pyplot as plt

# Imposto griglia chiara e trasparente in TUTTI i grafici
plt.rc('grid', color='lightgray', alpha=0.6)

#============  Funzione per caricare i dati  ============#
def load_data(path):
    data = np.loadtxt(path)
    blk   = data[:,0].astype(int)
    avg   = data[:,1]
    return blk, avg

#============  Funzione per plottare  ============#
def plot2(ax, blk, avg, ylabel, title):
    ax.plot(blk, avg, '-', color='slateblue', alpha=0.8, markersize=2)
    ax.set_xlabel("step")
    ax.set_ylabel(ylabel)
    ax.set_title(title)
    ax.grid(True)

#============  Caricamento dati  ============#
i_U1, r_U1 = load_data("terma_U1.txt")
i_U2, r_U2 = load_data("terma_U2.txt")
i_G1, r_G1 = load_data("terma_G1.txt")
i_G2, r_G2 = load_data("terma_G2.txt")

#============  Plot 4 subplot  ============#
fig, axs = plt.subplots(2, 2, figsize=(12, 8))
plt.suptitle("Termalizzazione Metropolis", fontsize=14)

plot2(axs[0,0], i_U1, r_U1, "r", "1s - Uniforme")
plot2(axs[1,0], i_U2, r_U2, "r", "2p - Uniforme")
plot2(axs[0,1], i_G1, r_G1, "r", "1s - Gaussiana")
plot2(axs[1,1], i_G2, r_G2, "r", "2p - Gaussiana")

plt.tight_layout(rect=[0,0,1,0.95])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Imposto griglia chiara e trasparente in TUTTI i grafici
plt.rc('grid', color='lightgray', alpha=0.6)

#============  Funzione per caricare i dati  ============#
def load_data(path):
    data = np.loadtxt(path)
    blk   = data[:,0].astype(int)  # step/blocco
    avg   = data[:,1]              # valor medio
    err   = data[:,2]              # errore
    return blk, avg, err

#============  Funzione per plottare con error bar ============#
def plot_error(ax, blk, avg, err, ylabel, title, expected):
    ax.errorbar(blk, avg, yerr=err, fmt='.', alpha=0.6, color='slateblue', ecolor='slateblue', markersize=5, elinewidth=1)
    
    # linea valore atteso
    ax.axhline(expected, linestyle='--', color='red', label=f"Valore atteso = {expected}")

    ax.set_xlabel("blocco")
    ax.set_ylabel(ylabel)
    ax.set_title(title)
    ax.legend()
    ax.grid(True)

#============  Caricamento dati  ============#
i_U1, r_U1, e_U1 = load_data("output_U1.txt")
i_U2, r_U2, e_U2 = load_data("output_U2.txt")
i_G1, r_G1, e_G1 = load_data("output_G1.txt")
i_G2, r_G2, e_G2 = load_data("output_G2.txt")

#============  Valori attesi (modificabili facilmente)  ============#
E_U1 = 1.5
E_U2 = 5.0
E_G1 = 1.5
E_G2 = 5.0

#============  Plot con barre d'errore  ============#
fig, axs = plt.subplots(2, 2, figsize=(12, 8))
plt.suptitle("Distanza media calcolata con Metropolis", fontsize=15)

plot_error(axs[0,0], i_U1, r_U1, e_U1, "⟨r⟩", "1s - Uniforme",  E_U1)
plot_error(axs[1,0], i_U2, r_U2, e_U2, "⟨r⟩", "2p - Uniforme",  E_U2)
plot_error(axs[0,1], i_G1, r_G1, e_G1, "⟨r⟩", "1s - Gaussiana", E_G1)
plot_error(axs[1,1], i_G2, r_G2, e_G2, "⟨r⟩", "2p - Gaussiana", E_G2)

plt.tight_layout(rect=[0,0,1,0.95])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Imposto griglia chiara e trasparente in TUTTI i grafici
plt.rc('grid', color='lightgray', alpha=0.6)

#============  Funzione per caricare i dati  ============#
def load_data(path):
    data = np.loadtxt(path)
    blk   = data[:,0].astype(int)
    err   = data[:,2]
    return blk, err

#============  Caricamento dati  ============#
i_U1, err_U1 = load_data("output_U1.txt")   # 1s Uniforme
i_U2, err_U2 = load_data("output_U2.txt")   # 2p Uniforme
i_G1, err_G1 = load_data("output_G1.txt")   # 1s Gaussiana
i_G2, err_G2 = load_data("output_G2.txt")   # 2p Gaussiana

#============  Plot a due pannelli  ============#
fig, axs = plt.subplots(1, 2, figsize=(12, 5))
plt.suptitle("Confronto dell'Errore", fontsize=15)

# ---- Plot 1s (Uniforme + Gaussiana) ---- #
axs[0].plot(i_U1, err_U1, label="1s Uniforme", color="royalblue", alpha=0.8)
axs[0].plot(i_G1, err_G1, label="1s Gaussiana", color="firebrick", alpha=0.8)
axs[0].set_title("Confronto 1s")
axs[0].set_xlabel("blocco")
axs[0].set_ylabel("errore")
axs[0].grid(True)
axs[0].legend()

# ---- Plot 2p (Uniforme + Gaussiana) ---- #
axs[1].plot(i_U2, err_U2, label="2p Uniforme", color="royalblue", alpha=0.8)
axs[1].plot(i_G2, err_G2, label="2p Gaussiana", color="firebrick", alpha=0.8)
axs[1].set_title("Confronto 2p")
axs[1].set_xlabel("blocco")
axs[1].set_ylabel("errore")
axs[1].grid(True)
axs[1].legend()

plt.tight_layout(rect=[0,0,1,0.93])
plt.show()
```
