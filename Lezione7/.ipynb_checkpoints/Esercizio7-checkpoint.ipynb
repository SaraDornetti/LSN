### Importance sampling usando l'algoritmo di Metropolis

Una limitazione dell'algoritmo di Metropolis è legata al fatto che spesso non $T(x|y)$ non contiene informazioni sulla densità di probabilità $p(x)$ che vogliamo campionare, di conseguenza vengono rifiutate più mosse del necessario. Ora supponiamo di poter scrivere $p(x)=d_1(x)d_2(x)$ e di saper campionare $d_1(x)$ direttamente (senza l'uso di $M(RT)^2$). Se scegliamo $T(x|y)=d_1(x)$, l'accettanza di $M(RT)^2$ diventa:

$$
A(x|y)=\min\left[1,\frac{d_2(x)}{d_2(y)}\right]
$$

Quindi, nel caso in cui $d_1(x)=p(x)$ vale $A(x|y)=1$, i.e. la mossa viene sempre accettata. L'idea è dunque quella di cercare di usare una $T(x|y)$ che siamo in grado di campionare, che sia il più possibile simile a $p(x)$.

Tuttavia è buona cosa correggere i risultati delle simulazioni compensando l'assenza del potenziale a lungo range. Per stimare questi contributi mancanti si assume che, per $r>r_{c}$ si ha $g(r)\approx 1$. Di conseguenza, prendendo come esempio il potenziale $V$: <span style="color:magenta">**cosa significa approssimare a 1? che sto considerando fuori da r_cut come disordinato??**</span>

$$
\langle V \rangle _{tail}=2\pi\rho N \int_{r_{c}}^{+\infty} dr r^2v(r)g(r) \approx 2\pi\rho N \int_{r_{c}}^{+\infty} dr r^2v(r)
$$

Per il potenziale di Lennard-Jones e per la pressione si ottiene:
$$
\frac{\langle V_{LJ} \rangle _{tail}}{N}= 8\pi\rho_{LJ}\varepsilon\left[\frac{1}{9 r_{c(LJ)}^9}-\frac{1}{3 r_{c(LJ)}^3}\right]
\hspace{3cm}
\frac{\langle w_{LJ} \rangle _{tail}}{3 N \varepsilon}= 32\pi\rho_{LJ}\left[\frac{1}{9 r_{c(LJ)}^9}-\frac{1}{6 r_{c(LJ)}^3}\right]
$$

## Tempo di correlazione
In un sistema complesso la memoria è l'abilità degli eventi passati di influenzare la dinamica futura, che si estende nel tempo in un modo invariante di scala. L'autocorrelazione di un processo random stazionario $x$ è la covarianza statistica del processo con una copia di se stesso nel futuro. L'autocorrelazione diventa uno strumento utile per studiare la presenza di memoria in un processo o, più precisamente, la dipendenza statistica in un processo.

Ciò di cui abbiamo bisogno è una misura del **tempo di correlazione** $t_c$ della simulazione. Il tempo di correlazione è una misura di quanto tempo il sistema impiega per passare da uno stato a un altro stato che è significativamente diverso dal primo. Per fare una stima si calcola la **funzione di autocorrelazione a tempi diversi** $\chi (t)$:

$$
\chi(t)=\frac{\langle m(t')m(t'+t)\rangle_{t'}-\langle m\rangle ^2}{\sigma_m^2}
$$

1. se $\chi(t)=0$  $\forall t>0$ i valori sono indipendenti e non c'è memoria.
2. se $\chi(t)=0$  $\forall t>t_c$ il processo diventa indipendente dalla storia passata dopo un certo tempo $t_c$, ovvero il tempo di correlazione.
3. se $\chi(t)\approx t^{-a}$ con $a>0$ e $t\gg 0$ il processo ha memoria a ogni tempo.


a grandi tempi ci aspettiamo che l'autocorrelazione abbia un andamento del tipo $\chi(t)\approx e^{-t/t_c}$. Se vogliamo campioni veramente indipendenti ci conviene prendere intervalli $\geq 2t_c$. Di conseguenza, se una simulazione procede per un tempo pari a $t_{max}$ (dopo aver fatto equilibrare il sistema per $t_{eq}$) possiamo prendere un numero di misure indipendenti pari a $n=t_{max}/2t_c$.

In una catena di Markov MC il tempo è discreto e la formula corretta per l'autocorrelazione diventa:

$$
\chi(t)=\frac{\frac{1}{t_{max}-t}\sum_{t'=0}^{t_{max}-t}m(t')m(t'+t)-\frac{1}{t_{max}-t}\sum_{t'=0}^{t_{max}-t}m(t')\times\frac{1}{t_{max}-t}\sum_{t'=0}^{t_{max}-t}m(t'+t)}{\frac{1}{t_{max}}\sum_{t'=0}^{t_{max}}m^2(t')-\left(\frac{1}{t_{max}}\sum_{t'=0}^{t_{max}}m(t')  \right)^2}
$$

Per vedere l'andamento dell'autocorrelazione, nel caso MC (NVT) partiamo da una configurazione a temperatura temp=1.1 e facciamo equilibrare il sistema per  $t_{eq}=5000 $ step. Per MD (NVE) partiamo da una temperatura superiore a quella di equilibrio, in modo tale da raggiungere la temperatura corretta dopo la fase di equilibrazione. Finita questa fase iniziale in cui si prepara il sistema si effettuano $M=5\times10^5$ step suddivisi in altrettanti blocchi di singolo step. In questo modo possiamo analizzare i dati grezzi da file tramite lo script di python. 

```python
import numpy as np
import matplotlib.pyplot as plt

# --- File: Energia Potenziale MD ---
data1 = np.loadtxt("NSL_SIMULATOR/OUTPUT_JN/MD/Autocorrelazione.dat")
step1 = data1[:, 0].astype(int)
penergy1 = data1[:, 1]

# --- File: Energia Potenziale MC ---
data2 = np.loadtxt("NSL_SIMULATOR/OUTPUT_JN/MC/Autocorrelazione.dat")
step2 = data2[:, 0].astype(int)
penergy2 = data2[:, 1]

def autocor(data, t_max):
    #Calcola la funzione di autocorrelazione normalizzata: χ(t) = [ <m(t')m(t'+t)> - <m(t')><m(t'+t)> ] / [ <m^2> - <m>^2 ] dove <...> denota la media su (t_max - t) punti. 
    N = len(data)
    if t_max > N:
        t_max = N

    # Varianza globale (denominatore)
    mean_tot = np.mean(data)
    var_tot = np.var(data)
    if var_tot == 0:
        raise ValueError("Varianza nulla, autocorrelazione non definita.")

    Chi = np.zeros(t_max)
    for t in range(t_max):
        M = N - t  # numero di coppie valide
        mean1 = np.mean(data[:M]) # <m(t')>
        mean2 = np.mean(data[t:M+t]) # <m(t'+t)>
        mean12 = np.mean(data[:M] * data[t:M+t])  # <m(t') m(t'+t)>
        Chi[t] = (mean12 - mean1 * mean2) / var_tot
    return Chi

t_max = 2000  # numero massimo di lag
Chi1 = autocor(penergy1, t_max)
Chi2 = autocor(penergy2, t_max)

fig, axes = plt.subplots(1, 2, figsize=(11,4), sharex=True)

axes[0].plot(np.arange(t_max), Chi1, color='slateblue')
axes[0].set_xlabel("Lag (t)")
axes[0].set_ylabel(r"$\chi_1(t)$")
axes[0].set_title("Autocorrelazione Simulazione MD")
axes[0].grid(alpha=0.5)

axes[1].plot(np.arange(t_max), Chi2, color='firebrick')
axes[1].set_xlabel("Lag (t)")
axes[1].set_ylabel(r"$\chi_2(t)$")
axes[1].set_title("Autocorrelazione Simulazione MC")
axes[1].grid(alpha=0.5)

plt.tight_layout()
plt.show()
```

## Data Blocking
Abbiamo $M$ misure di una certa quantità $g$ e le suddividiamo in $N$ blocchi, di conseguenza ogni blocco sarà costituito da $L=M/N$ misure. Il valore atteso si trova a partire dai valori medi nei singoli blocchi:
$$
\langle g \rangle = \frac{1}{N}\sum_{i=1}^Ng_i=\frac{1}{M}\sum_{i=1}^M g(\mu_i)
$$
L'errore sulla media è dato da:
$$
\sigma_{\langle g \rangle} =\sqrt{\frac{1}{N-1} \left(\langle g^2\rangle -\langle g \rangle ^2 \right)}
$$ 
dove $M$ è sostituito dal numero di blocchi.

```python
import numpy as np
import matplotlib.pyplot as plt
import math

# ===============================
# Funzioni
# ===============================

def block_error(data, L):
    """
    Calcola l'errore statistico stimato con il metodo dei blocchi per un dato L.
    data : array dei dati (es. energia potenziale)
    L    : dimensione del blocco
    """
    N = len(data) // L  # numero di blocchi interi
    if N < 2:
        return np.nan  # serve almeno 2 blocchi per stimare la varianza

    # Calcolo delle medie dei blocchi
    block_means = np.array([np.mean(data[i*L:(i+1)*L]) for i in range(N)])

    # Media progressiva e sua varianza
    mean_prog = np.cumsum(block_means) / np.arange(1, N + 1)
    mean2_prog = np.cumsum(block_means**2) / np.arange(1, N + 1)

    # Errore sul valor medio finale (ultimo valore)
    err = math.sqrt((mean2_prog[-1] - mean_prog[-1]**2) / (N - 1))
    return err


def compute_block_errors(data, L_values):
    """Calcola l'errore per ciascun valore di L."""
    return np.array([block_error(data, L) for L in L_values])


# ===============================
# Lettura dati
# ===============================

# Percorsi file (modifica se necessario)
file_MD = "NSL_SIMULATOR/OUTPUT_JN/MD/Autocorrelazione.dat"
file_MC = "NSL_SIMULATOR/OUTPUT_JN/MC/Autocorrelazione.dat"

# Caricamento dati
data_MD = np.loadtxt(file_MD)
data_MC = np.loadtxt(file_MC)

penergy_MD = data_MD[:, 1]
penergy_MC = data_MC[:, 1]

# ===============================
# Analisi
# ===============================

# Valori di L da esplorare (blocchi crescenti)
L_values = np.arange(10, 5000, 10)

# Calcolo errori
errors_MD = compute_block_errors(penergy_MD, L_values)
errors_MC = compute_block_errors(penergy_MC, L_values)

# ===============================
# Plot
# ===============================

fig, axes = plt.subplots(1, 2, figsize=(12, 4), sharey=True)

axes[0].plot(L_values, errors_MD, color='slateblue')
axes[0].set_xlabel("Larghezza del blocco L")
axes[0].set_ylabel("sigma")
axes[0].set_title("Dinamica Molecolare (MD)")
axes[0].grid(True, linestyle='--', alpha=0.6)

axes[1].plot(L_values, errors_MC, color='firebrick')
axes[1].set_xlabel("Larghezza del blocco L")
axes[1].set_title("Monte Carlo (MC)")
axes[1].grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

### Funzione di distribuzione radiale

la struttura di un fluido semplice monoatomico è caratterizzata da un set di funzioni di distribuzione per la posizione atomica, la più semplice è la  **funzione di distribuzione radiale** $g(r)$. Tale funzione restituisce la probabilità di trovare una coppia di atomi a distanza $r$ l'uno dall'altro, rispetto a un sistema disordinato che ha la stessa densità. Per definire 
$g(r)$, si parte dalle densità a $n$ particelle, che si ottengono integrando la funzione di distribuzione configurazionale sulle posizioni di tutti gli atomi tranne $n$, incorporando i fattori di normalizzazione appropriati:

$$
\rho_N^{(n)}(\vec{r}_1,\dots,\vec{r}_n)=\frac{N!}{(N-n)!}\int d\vec{r}_{n+1}\dots d\vec{r}_N\frac{e^{-\beta V(\vec{r}_1,\dots,\vec{r}_N)}}{Z} \hspace{1cm} \Rightarrow \hspace{1cm} \rho_N^{(2)}(ideal)= \rho\frac{N-1}{V}
$$

da cui:

$$
g_N^{(n)}(\vec{r}_1,\dots,\vec{r}_n)=\frac{\rho_N^{(n)}(\vec{r}_1,\dots,\vec{r}_n)}{\prod_{i=1}^n\rho_N^{(1)}(\vec{r}_i)} 
$$

La funzione di distribuzione radiale è utile non solo perchè contiene informazioni sulla struttura del liquido, ma anche perchè permette di calcolare le medie di insieme di qualsiasi funzione di coppia.

Ci serve un algoritmo per calcolare la funzione di distribuzione radiale. Per farlo osserviamo che per la densità a una particella:

$$
\langle \delta(\vec{r}-\vec{r}_1)\rangle = \int \delta(\vec{r}-\vec{r}_1) \frac{e^{-\beta V(\vec{r}_1,\dots,\vec{r}_N)}}{Z_N}d\vec{r}_{1}\dots d\vec{r}_N = \int \frac{e^{-\beta V(\vec{r},\vec{r}_2,\dots,\vec{r}_N)}}{Z_N}d\vec{r}_{2}\dots d\vec{r}_N \hspace{0.5cm} \Rightarrow \hspace{0.5cm} \rho_N^{(1)}(\vec{r})=\left\langle \sum_{i=1}^{N}\delta(\vec{r}-\vec{r}_i)\right\rangle
$$

A due particelle:

$$
\langle \delta(\vec{r}-\vec{r}_1) \delta(\vec{r}'-\vec{r}_2)\rangle = \int \delta(\vec{r}-\vec{r}_1)\delta(\vec{r}'-\vec{r}_2) \frac{e^{-\beta V(\vec{r}_1,\dots,\vec{r}_N)}}{Z_N}d\vec{r}_{1}\dots d\vec{r}_N = \int \frac{e^{-\beta V(\vec{r},\vec{r}',\vec{r}_3,\dots,\vec{r}_N)}}{Z_N}d\vec{r}_{3}\dots d\vec{r}_N \hspace{0.5cm} \Rightarrow \hspace{0.5cm} \rho_N^{(2)}(\vec{r})=\left\langle \sum\sum_{i\neq j=1}^{N}\delta(\vec{r}-\vec{r}_i)\delta(\vec{r}'-\vec{r}_j)\right\rangle
$$

In un **liquido omogeneo**, in cui la funzione di distribuzione radiale è solo funzione della differenza tra i vettori $\vec{r}_1$ e $\vec{r}_2$, segue che $g^{(2)}(\vec{r}_1,\vec{r}_2)=g(r)$. In questo caso i conti sono un po' più semplici e si arriva a:

$$
g(r)=\langle g(r)\rangle = \frac{1}{\rho N \Delta V(r)}\left\langle \sum \sum_{i\neq j=1}^N \delta  (|\vec{r}|-|(\vec{r}_i-\vec{r}_j)| )\right\rangle
$$

Dove $V(r)=\frac{4\pi}{3}[(r+dr)^3-r^3]$ e la media è calcolata su:
- $r\in[r;r+dr]$
- $\theta\in[0;\pi]$
- $\phi\in[0;2\pi]$

Durante la simulazione bisogna riempire un istogramma aumentando il bin ... ogni volta che si trovano due particelle a distanza inferiore di $r+dr$. Alla fine della simulazione si normalizza l'istogramma dividendo per $\rho N \Delta V(r)$.

Per calcolare correttamente le pressioni si modifica system.cpp: per MC si usa il valore di temperatura `_temp`, per MD bisogna calcolare di volta in volta la temperatura a partire dall'energia cinetica calcolata per la configurazione attuale.

```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 3, figsize=(18,5))

# =====================================================
# 1) Energia Potenziale
# =====================================================
data_md = np.loadtxt("NSL_SIMULATOR/OUTPUT_JN/MD/potential_energy.dat")
data_mc = np.loadtxt("NSL_SIMULATOR/OUTPUT_JN/MC/potential_energy.dat")

blk_md   = data_md[:, 0]
media_md = data_md[:, 2]
err_md   = data_md[:, 3]

blk_mc   = data_mc[:, 0]
media_mc = data_mc[:, 2]
err_mc   = data_mc[:, 3]

ax[0].errorbar(blk_md, media_md, yerr=err_md, fmt='.-', alpha=0.6, color='slateblue', markersize=5, label="MD")
ax[0].errorbar(blk_mc, media_mc, yerr=err_mc, fmt='.-', alpha=0.6, color='firebrick', markersize=5, label="MC")
ax[0].set_xlabel("blocco")
ax[0].set_ylabel("Energia potenziale")
ax[0].set_title("Energia potenziale")
ax[0].grid(True,alpha=0.5)
ax[0].legend()

# =====================================================
# 2) Pressione
# =====================================================
data_md = np.loadtxt("NSL_SIMULATOR/OUTPUT_JN/MD/pressure.dat")
data_mc = np.loadtxt("NSL_SIMULATOR/OUTPUT_JN/MC/pressure.dat")

blk_md   = data_md[:, 0]
media_md = data_md[:, 2]
err_md   = data_md[:, 3]

blk_mc   = data_mc[:, 0]
media_mc = data_mc[:, 2]
err_mc   = data_mc[:, 3]

ax[1].errorbar(blk_md, media_md, yerr=err_md, fmt='.-', alpha=0.6, color='slateblue', markersize=5, label="MD")
ax[1].errorbar(blk_mc, media_mc, yerr=err_mc, fmt='.-', alpha=0.6, color='firebrick', markersize=5, label="MC")
ax[1].set_xlabel("blocco")
ax[1].set_ylabel("pressione")
ax[1].set_title("pressione")
ax[1].grid(True,alpha=0.5)
ax[1].legend()

# =====================================================
# 3) g(r)
# =====================================================
data_md = np.loadtxt("NSL_SIMULATOR/OUTPUT_JN/MD/gofr.dat")
data_mc = np.loadtxt("NSL_SIMULATOR/OUTPUT_JN/MC/gofr.dat")

blk_md   = data_md[:,0]
r_md     = data_md[:,1]
cum_md   = data_md[:,3]
err_md   = data_md[:,4]

blk_mc   = data_mc[:,0]
r_mc     = data_mc[:,1]
cum_mc   = data_mc[:,3]
err_mc   = data_mc[:,4]

mask_md = blk_md == blk_md.max()
mask_mc = blk_mc == blk_mc.max()

ax[2].errorbar(r_md[mask_md], cum_md[mask_md], yerr=err_md[mask_md], fmt='*-', color='slateblue', markersize=5, label="g(r) MD")
ax[2].errorbar(r_mc[mask_mc], cum_mc[mask_mc], yerr=err_mc[mask_mc], fmt='.-', color='firebrick', markersize=3, label="g(r) MC")
ax[2].set_xlabel("r")
ax[2].set_ylabel("g(r)")
ax[2].set_title("g(r)")
ax[2].grid(True,alpha=0.5)
ax[2].legend()


plt.tight_layout()
plt.show()
```

Nelle simulazioni svolte sono stati scelti i parametri iniziali con molta cura per soddisfare le condizioni iniziali richieste post equilibrazione. Partendo da configurazioni iniziali poco diverse (ad esempio $T_{start}=2$ per MD) si otterrebbe una temperatura media all'equilibrio spostata rispetto al valore atteso e questo sarebbe sufficiente per ottenere risultati di energia e pressione significativamente diversi rispetto a quelli ottenuti tramite simulazione MC. Queste differenze sono spiegate dal fatto che se equilibriamo confrontiamo sistemi diversi. In generale delle discrepanze tra i due modelli sono non sono preoccupanti perchè si tratta di due algoritmi completamente diversi.
